```python
import csv
import pandas as pd
import time
from selenium import webdriver
from operator import itemgetter, attrgetter
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://movie.douban.com/tag/#/?sort=S&range=1,10&tags=%E7%94%B5%E8%A7%86%E5%89%A7,%E4%B8%AD%E5%9B%BD%E5%A4%A7%E9%99%86,2010%E5%B9%B4%E4%BB%A3&playable=1'
driver = webdriver.Chrome()
driver.get(url)
```

```python
#点击“加载更多”方法一：
def one_click():
    js = "var q = document.documentElement.scrollTop = 10000"
    driver.execute_script(js)
    driver.find_element_by_class_name('more').click()
    time.sleep(2)
```

```python
while True:
    try:
        one_click()
    except:
        break
```

```python
driver.find_element_by_class_name('th-list').click()
```

```python
# scores = []
# for k in driver.find_elements_by_class_name('rating'):
#     score = k.text
#     scores.append(score)

#简写
scores = [k.text for k in driver.find_elements_by_class_name('rating')]
directors_and_actors = [k.text for k in driver.find_elements_by_class_name('cast')]
titles = [k.text for k in driver.find_elements_by_class_name('title')]
```

```python
# url = []
# h = driver.find_element_by_css_selector('.item a')
# h.get_attribute('href')

#简写
url = [k.get_attribute('href') for k in driver.find_elements_by_class_name('item')]
```

```python
mylist = []
for i in range(len(scores)):
    if directors_and_actors[i].find('导演：') != -1 and directors_and_actors[i].find('主演：') != -1:
    #去除没有导演和主演的电视剧
        dir_and_act = directors_and_actors[i].split('\n')
        mylist.append([titles[i],scores[i],dir_and_act[0].lstrip('导演：'),dir_and_act[1].lstrip('主演：'),url[i]])
        #去除“导演：”和“主演：”
```

```python
mylist
```

```python
#豆瓣原本按照评分的排序有错，因此需要对这个list重新排序
#方法一：
#my_new_lists = sorted(mylists, key = lambda a:a[1], reverse = True)
```

```python
#方法二：
my_new_lists = sorted(mylist, key = itemgetter(1), reverse = True)
```

```python
with open('my_TV_series_list.csv','w',encoding = 'gbk') as f:
    mywriter = csv.writer(f)
    mywriter.writerow(['Title','Score','Directors','Actors','url'])
    mywriter.writerows(mylist)
```
