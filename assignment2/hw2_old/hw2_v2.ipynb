```python
import pandas as pd
from dateutil.parser import parse
import matplotlib
%matplotlib inline
import plotly  
from matplotlib import pyplot as plt
from pyecharts import Bar
from pyecharts import Line
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('results.csv',encoding = 'gbk')
```

```python
df['年份'] = df['年份'].apply(int)
```

```python
year2010 = df[(df['年份']<2011)]
year2011 = df[(df['年份']>2010) & (df['年份']<2012)]
year2012 = df[(df['年份']>2011) & (df['年份']<2013)]
year2013 = df[(df['年份']>2012) & (df['年份']<2014)]
year2014 = df[(df['年份']>2013) & (df['年份']<2015)]
year2015 = df[(df['年份']>2014) & (df['年份']<2016)]
year2016 = df[(df['年份']>2015) & (df['年份']<2017)]
year2017 = df[(df['年份']>2016) & (df['年份']<2018)]
year2018 = df[(df['年份']>2017)]
```

```python
df_year = pd.DataFrame({'年份':['2010',
                             '2011',
                             '2012',
                             '2013',
                             '2014',
                             '2015',
                             '2016',
                             '2017',
                             '2018',
                            ],
                        '评价总人数':[year2010['评价人数'].sum(),
                                year2011['评价人数'].sum(),
                                year2012['评价人数'].sum(),
                                year2013['评价人数'].sum(),
                                year2014['评价人数'].sum(),
                                year2015['评价人数'].sum(),
                                year2016['评价人数'].sum(),
                                year2017['评价人数'].sum(),
                                year2018['评价人数'].sum()
                               ]})
```

```python
styles = []
for i in df['类型'].fillna(''):
    style = i.strip().split(' / ')
    for j in style:
        styles.append(j)

tmp = pd.DataFrame({'styles': styles})
tmp = tmp[tmp['styles'] != '']
```

```python
from wordcloud import WordCloud
from matplotlib import pyplot as plt
%matplotlib inline
x = tmp['styles'].value_counts()
wc = WordCloud(background_color = 'white', font_path = 'SimHei.ttf', max_words = 2000, max_font_size = 100, random_state = 30)
wc.generate_from_frequencies(x)
plt.figure(figsize=(14, 7))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```
