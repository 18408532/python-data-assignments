```python
import pandas as pd
import matplotlib

%matplotlib inline
```

```python
df = pd.read_csv('results.csv', encoding = 'gbk')
```

```python
from dateutil.parser import parse
```

```python
styles = []
for i in df['类型'].fillna(''):
    style = i.strip().split(' / ')
    for j in style:
        styles.append(j)
```

```python
styles
```

```python
tmp = pd.DataFrame({'styles': styles})
```

```python
tmp = tmp[tmp['styles'] != '']
```

```python
del tmp['times']
```

```python
from wordcloud import WordCloud
from matplotlib import pyplot as plt
x = tmp['styles'].value_counts()
wc = WordCloud(background_color = 'white', font_path = 'msyh.ttf', max_words = 2000, max_font_size = 100, random_state = 20)
wc.generate_from_frequencies(x)
plt.figure(figsize=(14, 7))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```
